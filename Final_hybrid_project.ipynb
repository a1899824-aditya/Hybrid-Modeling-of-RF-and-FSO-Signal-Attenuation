```python
import pandas as pd
import numpy as np
from sklearn.utils import resample

print("\n--- Performing Dataset Downsampling & Balancing ---")

# Step 1: Display the initial SYNOPCode distribution
synop_distribution = df["SYNOPCode"].value_counts()
print("Initial SYNOPCode Distribution:\n", synop_distribution)

# Step 2: Define Target Sample Size per SYNOP Code
median_count = int(synop_distribution.median())  # Using the median count
max_limit = int(1.5 * median_count)  # Capping max limit to avoid over-sampling

print(f"\nTarget Records Per SYNOPCode: {median_count} (Capped at {max_limit} for majority classes)")

# Step 3: Create a balanced dataset
balanced_df = pd.DataFrame(columns=df.columns)

for synop_code, count in synop_distribution.items():
    subset = df[df["SYNOPCode"] == synop_code]
    
    if count >= max_limit:
        # Downsample majority classes
        subset_balanced = subset.sample(max_limit, random_state=42)
    elif count >= median_count:
        # Keep as is if within range
        subset_balanced = subset
    else:
        # Bootstrap minority classes
        subset_balanced = resample(subset, replace=True, n_samples=median_count, random_state=42)
    
    balanced_df = pd.concat([balanced_df, subset_balanced], ignore_index=True)

# Step 4: Verify the new SYNOPCode distribution
balanced_distribution = balanced_df["SYNOPCode"].value_counts()
print("\nBalanced SYNOPCode Distribution:\n", balanced_distribution)

# Step 5: Save the balanced dataset
balanced_df.to_csv("Balanced_Dataset.csv", index=False)
print("\nBalanced dataset saved as 'Balanced_Dataset.csv'.")
```

```python
# âœ… Import Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# âœ… Load Dataset
df = pd.read_csv("Balanced_Dataset.csv")

# âœ… Get Unique SYNOP Codes
synop_codes = df["SYNOPCode"].unique()

# âœ… Define Backward Feature Elimination Function
def backward_feature_elimination(X, y, synop_code, target_name):
    """
    Perform backward feature elimination using Out-Of-Bag (OOB) score & RMSE.
    """
    model = RandomForestRegressor(n_estimators=100, oob_score=True, random_state=42)
    features = list(X.columns)
    best_features = features.copy()
    best_rmse = float("inf")
    best_r2 = float("-inf")
    
    rmse_values, r2_values, removed_features, num_features, oob_scores = [], [], [], [], []

    while len(features) > 1:
        # Split Data
        X_train, X_test, y_train, y_test = train_test_split(X[features], y, test_size=0.2, random_state=42)
        
        # Train Model
        model.fit(X_train, y_train)
        
        # Predict & Compute Performance
        y_pred = model.predict(X_test)
        current_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
        current_r2 = r2_score(y_test, y_pred)
        current_oob = model.oob_score_
        
        # Identify the Least Important Feature
        importances = model.feature_importances_
        least_important = features[np.argmin(importances)]
        
        # Store Performance at Each Step
        rmse_values.append(current_rmse)
        r2_values.append(current_r2)
        removed_features.append(least_important)
        num_features.append(len(features))
        oob_scores.append(current_oob)
        
        # Update Best Features Set
        if current_rmse < best_rmse:
            best_rmse = current_rmse
            best_r2 = current_r2
            best_features = features.copy()
        
        # Remove Least Important Feature
        features.remove(least_important)
    
    # Save Selected Features
    selected_features_df = pd.DataFrame({"Selected Features": best_features})
    selected_features_df.to_csv(f"selected_features_{target_name}_SYNOP_{synop_code}.csv", index=False)
    
    # Save Feature Elimination Summary
    elimination_df = pd.DataFrame({
        "Step": range(1, len(removed_features) + 1),
        "Removed Feature": removed_features,
        "Remaining Features": num_features,
        "RMSE": rmse_values,
        "R2 Score": r2_values,
        "OOB Score": oob_scores
    })
    elimination_df.to_csv(f"feature_elimination_{target_name}_SYNOP_{synop_code}.csv", index=False)
    
    return best_features, elimination_df

# âœ… Perform Feature Selection for Each SYNOP Code
selected_features_per_synop = {}

for synop_code in synop_codes:
    df_synop = df[df["SYNOPCode"] == synop_code]
    
    # Skip if there are too few samples
    if len(df_synop) < 50:
        continue
    
    # Define Features & Targets
    X_rfl = df_synop.drop(columns=["FSO_Att", "RFL_Att", "SYNOPCode"])  # Remove target & categorical column
    y_rfl = df_synop["RFL_Att"]
    
    # Perform Feature Selection for RFL_Att
    best_rfl_features, rfl_elimination_df = backward_feature_elimination(X_rfl, y_rfl, synop_code, "RFL_Att")
    
    # Store selected features
    selected_features_per_synop[synop_code] = best_rfl_features

print("\nâœ… Feature Selection Complete! Selected Features Saved Per SYNOP Code.")
```

```python
# âœ… Initialize Storage for Model Performance
synop_model_results = []

# âœ… Train SYNOP-Specific RF Models
for synop_code in synop_codes:
    df_synop = df[df["SYNOPCode"] == synop_code]
    
    # Skip if there are too few samples
    if len(df_synop) < 50:
        continue
    
    # Load Selected Features
    selected_features = selected_features_per_synop[synop_code]
    
    # Define Features & Targets
    X_rfl = df_synop[selected_features]
    y_rfl = df_synop["RFL_Att"]
    
    # Train-Test Split
    X_train, X_test, y_train, y_test = train_test_split(X_rfl, y_rfl, test_size=0.2, random_state=42)
    
    # Train Random Forest Model
    rf_model = RandomForestRegressor(n_estimators=100, oob_score=True, random_state=42)
    rf_model.fit(X_train, y_train)
    
    # Evaluate Model
    y_pred = rf_model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    oob = rf_model.oob_score_
    
    # Store Results
    synop_model_results.append({
        "SYNOPCode": synop_code,
        "RMSE": rmse,
        "R2 Score": r2,
        "OOB Score": oob
    })

# âœ… Convert to DataFrame & Save
synop_results_df = pd.DataFrame(synop_model_results)
synop_results_df.to_csv("SYNOP_RFL_Model_Results.csv", index=False)

print("\nâœ… SYNOP-Specific Model Training Complete! Results Saved.")
```
